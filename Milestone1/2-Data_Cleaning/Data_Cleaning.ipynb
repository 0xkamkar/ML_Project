```python
import numpy as np
import pandas as pd
```

```python
all_podcasts = pd.read_csv('all_podcasts_df.csv').drop(['Unnamed: 0'], axis=1)
all_podcasts
```

```python
my_podcaster = pd.read_csv('my_podcaster_df.csv').drop(['Unnamed: 0'], axis=1)
my_podcaster
```

<div dir=rtl >
<font  size=3>
از دیتافریم all_podcasts تنها پادکست‌هایی را نگه می‌داریم که podcaster_id آن در دیتافریم my_podcaster باشد.

```python
my_podcasts1 = all_podcasts.merge(my_podcaster, on='podcaster_id')
my_podcasts1
```

<div dir=rtl >
<font  size=3>
همانطور که مشخص است در دیتافریم بالا بعضی پادکست‌های مشابه تکرار شده‌اند. اطلاعاتی مانند اسم پادکست، تعداد قسمت‌های آن و اطلاعات مربوط به پادکستر مشابه است ولی cid و sub_coun متفاوت است. با بررسی پادکست‌های تکراری به این نتیجه رسیدیم که بین پادکست‌هایی که عنوان و پادکستر مشابه دارند پادکستی را نگه داریم که بیشترین sub_count را دارد.

```python
my_podcasts2 = my_podcasts1.sort_values(['sub_count'], ascending=False)
my_podcasts3 = my_podcasts2.drop_duplicates(['title', 'podcaster_id'])
my_podcasts3
```

```python
my_podcasts3.to_csv('my_podcasts_df.csv')
```
